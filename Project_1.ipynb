```python
 sns.countplot(data=ds,x='Age Group', hue='Gender')
plt.show()
```

```python
#total numbers of order/Top 10 states
sales_states  = ds.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_states, x = 'State',y= 'Orders')
plt.show()
```

```python
#Total amount/sales of top status
sales_state= ds.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

In this graph the Uttar Pradesh have a highest ranking in orders
